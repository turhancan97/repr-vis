```python
import os
import shutil

def copy_and_fix_json(src_dir, dst_dir):
    for root, dirs, files in os.walk(src_dir):
        for file in files:
            if file.lower().endswith('.json'):
                src_path = os.path.join(root, file)

                # Preserve relative directory structure
                rel_path = os.path.relpath(root, src_dir)
                dst_folder = os.path.join(dst_dir, rel_path)
                os.makedirs(dst_folder, exist_ok=True)

                dst_path = os.path.join(dst_folder, file)

                # Read, replace, write
                with open(src_path, 'r', encoding='utf-8') as f:
                    content = f.read()

                content = content.replace('.png', '.jpg')

                with open(dst_path, 'w', encoding='utf-8') as f:
                    f.write(content)

                print(f"Copied and fixed: {src_path} -> {dst_path}")
                
copy_and_fix_json("Frames_Line", "Frames_Line_jpg")
copy_and_fix_json("Frames_Orbit", "Frames_Orbit_jpg")
```
